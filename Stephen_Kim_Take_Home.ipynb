```python
#Ratio of Employees/Contractors
df['CNA_cont_ratio'] = df['Hrs_CNA_ctr'] / (df['Hrs_CNA_emp'] / df['Hrs_CNA_ctr']).replace({0: pd.NA})
df['CNA_cont_ratio']

df['LPN_cont_ratio'] = df['Hrs_LPN_ctr'] / (df['Hrs_LPN_emp'] / df['Hrs_LPN_ctr']).replace({0: pd.NA})
df['LPN_cont_ratio']

df['RN_cont_ratio'] = df['Hrs_RN_ctr'] / (df['Hrs_RN_emp'] / df['Hrs_RN_ctr']).replace({0: pd.NA})
df['RN_cont_ratio']
```

```python
df['LPN_cont_ratio'].value_counts()
```

```python
df['RN_cont_ratio'].value_counts()
df['CNA_cont_ratio'].value_counts()
```

```python
lpn_mean_ratio = df['LPN_cont_ratio'].mean()

# Group the DataFrame by 'PROVNAME' and calculate the mean 'CNA_cont_ratio' for each group
grouped_resultLPN = df.groupby('PROVNAME')[['LPN_cont_ratio']].mean().reset_index()

# Filter the grouped DataFrame to include only rows where 'CNA_cont_ratio' is greater than the overall mean
high_lpn_contractor_facilities = grouped_resultLPN[grouped_resultLPN['LPN_cont_ratio'] > lpn_mean_ratio].sort_values(by='LPN_cont_ratio', ascending=False)

# Display the result
high_lpn_contractor_facilities
```

```python
cna_mean_ratio = df['CNA_cont_ratio'].mean()

# Group the DataFrame by 'PROVNAME' and calculate the mean 'CNA_cont_ratio' for each group
grouped_result = df.groupby('PROVNAME')['CNA_cont_ratio'].mean().reset_index()

# Filter the grouped DataFrame to include only rows where 'CNA_cont_ratio' is greater than the overall mean
high_cna_contractor_facilities = grouped_result[grouped_result['CNA_cont_ratio'] > cna_mean_ratio]

# Display the result
high_cna_contractor_facilities
```
